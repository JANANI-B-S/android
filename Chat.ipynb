```python
from transformers import MarianMTModel, MarianTokenizer
import langid

# Supported language codes by the Helsinki-NLP models
language_codes = {
    'en': 'English', 'fr': 'French', 'de': 'German', 'es': 'Spanish', 'it': 'Italian',
    'ru': 'Russian', 'zh': 'Chinese', 'ja': 'Japanese', 'ko': 'Korean', 'pt': 'Portuguese',
    'ar': 'Arabic', 'hi': 'Hindi', 'bn': 'Bengali', 'tr': 'Turkish', 'vi': 'Vietnamese'
}

def translate_text(text, tgt_lang):
    # Detect the source language
    src_lang, _ = langid.classify(text)

    if src_lang not in language_codes or tgt_lang not in language_codes:
        return "Translation not supported for this language pair."

    model_name = f'Helsinki-NLP/opus-mt-{src_lang}-{tgt_lang}'

    try:
        # Load model and tokenizer
        model = MarianMTModel.from_pretrained(model_name)
        tokenizer = MarianTokenizer.from_pretrained(model_name)
    except Exception as e:
        return f"Model loading failed: {str(e)}"

    # Tokenize the text using the recommended method
    model_inputs = tokenizer(text, return_tensors="pt")

    # Perform the translation
    translated_tokens = model.generate(**model_inputs)
    translated_text = tokenizer.batch_decode(translated_tokens, skip_special_tokens=True)[0]

    return translated_text

# User input
text_to_translate = input("Enter the text you want to translate: ")
print("Available languages: ", list(language_codes.values()))
target_lang = input("Enter the target language code (e.g., 'fr' for French): ")

# Translate and display the result
translated_text = translate_text(text_to_translate, target_lang)
print(f"Translated text: {translated_text}")

```
